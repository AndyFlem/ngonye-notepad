```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("power_energy.csv")
data['Date']=pd.to_datetime(data['Date'],format="%d/%m/%Y")
data=data.set_index(pd.DatetimeIndex(data['Date']))
data=data.drop(['Date'],axis=1)

data['Year']=data.index.year
data['Month']=data.index.month
data['Day']=data.index.day

data['MonthId']=data['Year']+data['Month']/100
data['WaterYear']=data.apply((lambda x: (x['Year'] if x['Month']>=10 else x['Year']-1)),axis=1)
data['WaterMonth']=data.apply((lambda x: (x['Month']-9 if x['Month']>=10 else x['Month']+3)),axis=1)
data['EFlow_Prop']=data['Total_EFlow']/data['Inflow']

data.info()
```

```python
monthly=data.groupby(['Month','MonthId','WaterYear','WaterMonth'], as_index=False).aggregate(np.mean).set_index('MonthId')
monthly['Energy']=data[['MonthId','Energy']].groupby('MonthId').aggregate('sum')
monthly=monthly.sort_values(by=['MonthId'])
monthly
```

```python
yearly=monthly[['WaterYear','Energy']].groupby(['WaterYear']).aggregate('sum')
yearly
```

```python
mean_annual_energy=yearly['Energy'].mean()
mean_annual_energy
```

```python
yearly['Energy_pct_var']=(yearly['Energy']-mean_annual_energy)/mean_annual_energy*100
yearly
```
